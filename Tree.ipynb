# Tree Traversal 
Tree Traversal is basically going through every node(key) exactly once.
- Necessary to search for an item
- Size of a tree

Tree Traversals by BFS or DFS
BFS - Breadth First 
DFS - Depth First Traversal 

DFT is like recursive - Traverse Root, Traverse Left Subtree and Traverse Right Subtree   
3! total permutations  
3 most famous are - Inorder, Preorder and Postorder  

Inorder - L Ro R  
Preorder - Ro L R  
PostOrder - L R Ro

```python
# Inorder Traversal 
# Time Complexity - Theta(n) n is number of nodes as each node is called once for inorder. n traverses calls 
# for each node and then constant work for each node
# Auxillary Space - Theta(h) h is the height of tree

class Node:
    def __init__(self, k):
        self.left = None
        self.right = None
        self.key = k


def inorder(root):
    if root != None: #check whether the tree is empty
        inorder(root.left)
        print(root.key)
        inorder(root.right)


# Driver Code

root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.right.left = Node(40)
root.right.right = Node(50)

inorder(root)
```

```python
## Size of the Binary Tree - Number of Nodes 
# When we are given a binary tree, we refer to root of the binary tree is given
# If root is None then return 0
#    - Handles empty tree
#    - handles leaf nodes

class Node:
    def __init__(self, k):
        self.left = None
        self.right = None
        self.key = k


def treeSize(root):
    if root == None:
        return 0
    else:
        ls = treeSize(root.left)
        rs = treeSize(root.right)
        return ls + rs + 1
        # return 1 + treeSize(root.left) + treeSize(root.right)

# Driver Code

root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.right.left = Node(40)
root.right.right = Node(50)

print(treeSize(root))

# Time Complexity - Theta(n)
# Auxillary Space - Depends on recursion call stack - Theta(h)
# Height of tree is number of nodes from root to leaf
```

```python
## Maximum in Binary Tree

from math import inf


class Node:
    def __init__(self, k):
        self.left = None
        self.right = None
        self.key = k


def getMax(root):
    if root == None:
        return -inf
    else:
        lm = getMax(root.left)
        rm = getMax(root.right)

        return max(root.key, lm, rm)


# Driver Code

root = Node(10)
root.left = Node(80)
root.right = Node(15)
root.right.left = Node(40)
root.right.right = Node(50)

print(getMax(root))

#Same as Theta(n) and Theta(h)
```

```python
## Search in the Binary Tree

class Node:
    def __init__(self, k):
        self.left = None
        self.right = None
        self.key = k


def searchkey(root, key):
    if root is None:
        return False
    elif root.key == key:
        return True
    elif searchkey(root.left, key):
        return True
    else:
        return searchkey(root.right, key)


# Driver Code

root = Node(10)
root.left = Node(12)
root.right = Node(40)
root.right.left = Node(15)
root.right.right = Node(30)

print(searchkey(root, 15))
print(searchkey(root, 25))
print(searchkey(root, 30))

# Time Complexity - O(n)
# Auxillary Space - O(h)
```

```python
## Height of the Binary Tree
# Max no of nodes from Root to Leaf path

class Node:
    def __init__(self, k):
        self.left = None
        self.right = None
        self.key = k

def height(root):
    if root == None:
        return 0
    else:
        lh = height(root.left)
        rh = height(root.right)
        return max(lh, rh) + 1


# Driver Code

root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.right.left = Node(40)

print(height(root))
```

```python
## Iterative Order In Order Travesal without Recursion

class Node:
    def __init__(self, k):
        self.left = None
        self.right = None
        self.key = k


def itrInorder(root):
    if root is None:
        return

    st = []
    curr = root
    while curr is not None:
        st.append(curr)
        curr = curr.left

    while len(st) > 0:
        curr = st.pop()
        print(curr.key)
        curr = curr.right

        while curr is not None:
            st.append(curr)
            curr = curr.left


# Driver Code

root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.left.left = Node(40)
root.left.right = Node(50)

itrInorder(root)
```

```python
## Iterative Pre Order Traversal of Tree

class Node:
    def __init__(self, k):
        self.left = None
        self.right = None
        self.key = k


def itrPreorder(root):
    if root is None:
        return

    st = [root]

    while len(st) > 0:
        curr = st.pop()
        print(curr.key)

        if curr.right is not None:
            st.append(curr.right)

        if curr.left is not None:
            st.append(curr.left)


# Driver Code

root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.left.left = Node(40)
root.left.right = Node(50)
root.right.right = Node(60)

itrPreorder(root)
```

```python
# BST similar to tree
# For every node, keys in left side are smaller, and keys in right are greater
# All keys are distinct 
# Linked Data Structure
# Search in BST is simple O(logN) - every time removing half of the tree
# O(n) for BInary Tree but O(logN) for BST
# Create an empty BST - same as normal BT
# Insert keys one by one 
```

```python
# Search in BST (Root and Key are given)
# Recursive

def searchBst(root, key):
    if root is None:
        return False

    elif root.key == key:
        return True
    elif root.key > key:
        return searchBst(root.left, key)
    else:
        return searchBst(root.right, key)
```

```python
# Iteratively

def searchBst(root, key):
    while root is not None:

        if root.key == key:
            return True
        elif root.key > key:
            return searchBst(root.left, key)
        else:
            return searchBst(root.right, key)

    return False
```

```python
## BST Insert in Python
# Insertion always happen at leaf
# Insert should be assigned to left or right 
# Always return root

class Node:
    def __init__(self, key):
        self.left = None
        self.key = key
        self.right = None


def insert(root, key):
    if root == None:
        return Node(key)
    elif root.key == key:
        return root # No insertion as key already present
    elif root.key > key:
        root.left = insert(root.left, key)
    else:
        root.right = insert(root.right, key)
    return root
```

```python
# Delete key fom BST
# Inorder traversal of BST is always sorted
# Inorder Successor is closet greater value to be deleted
# 

class Node:
    def __init__(self, key):
        self.left = None
        self.key = key
        self.right = None



def getSucc(curr, key):
    while curr.left != None:
        curr = curr.left

    return curr.key


def deleteNode(root, key):
    if root == None:
        return

    if root.key > key:
        root.left = deleteNode(root.left, key)

    if root.key < key:
        root.right = deleteNode(root.right, key)

    else:
        if root.left == None:
            return root.right
        elif root.right == None:
            return root.left
        else:
            succ = getSucc(root.right, key)
            root.key = succ
            root.right = deleteNode(root.right, succ)

    return root

# Time Complexity - O(h)
# Auxillary Space - O(h)
```
